```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(0)

X = rng.normal(0, 1, size=(1000, 2))
x, y = X[:, 0], X[:, 1]
z = np.arange(4).repeat(x.size / 4)
#z = np.abs(x) + np.abs(y) #np.linspace(0, 10, x.size)
```

```python
sns_theme = {
    "context": "paper",
    "style": "ticks",
    "palette": "deep",
    "font": "serif",
}

sns.set_theme(**sns_theme)

rcp = {
    "figure.dpi": 50,
}
plt.rcParams.update(rcp)
```

```python
def set_height(width, ratio=1, fraction=1, subplots=(1, 1)):

    width_per_subplot = fraction * width / subplots[1]
    height_per_subplot = ratio * width_per_subplot

    return {"height": height_per_subplot, "aspect": ratio}
```

```python
cmap = "viridis"
ax = sns.relplot(x=x, y=y, **set_height(3, 2))# hue=z)
#ax._legend.remove()

norm = plt.Normalize(z.min(), z.max())
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
#ax.figure.colorbar(sm, label="Z")

ax.set(xlabel="$\sigma$", ylabel="Y")

#ax.fig.set_size_inches(5, 5)
ax.despine(right=False)
plt.show()
```
